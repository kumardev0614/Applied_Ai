```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv("iris.csv")
```

# Univariate Analysis
Here we will analyse each variable to see how efficent the variable is to be used in model.

```python
# The rule of thumb is that "The more the seperated PDFs the better our variable for classification"
# More overlap of PDF lines means the bad our variable is for classification.
```

### sepal_width.

```python
sns.distplot(iris.sepal_width)

# As you can see this code has created a distplot of all 150 sepal_width values.
# But we need to see the distribution with respect to species.
# To create 3 distplots from 50 values of setosa, 50 values of versicolor and 50 values of virgina on a single plot.
# We have to use sns.FacetGrid. Where hue='species' can be applied.
```

```python
sns.FacetGrid(iris, hue='species', height=5).map(sns.distplot, 'sepal_width').add_legend()
plt.show()
```

```python
# sw is will be very bad if we have to choose it.
# Because there is no seperation and overlaping is too much.
```

### sepal_length

```python
sns.FacetGrid(iris, hue='species', height=5).map(sns.distplot, 'sepal_length').add_legend()
plt.show()
```

```python
v = iris.loc[ iris.species == 'virginica' ]
```

```python
print(list(v.sepal_length))
```

```python
v.loc[ (v.sepal_length >= 7) & (v.sepal_length < 8) ]
```

```python
v.loc[ (v.sepal_length >= 7) & (v.sepal_length < 8) ].sepal_length.count()
```

```python
# Sl is better than SW Because overlap is lesser than SW. 
# But still it will be a bad choice beacuse there is considrable overlap.
```

### petal_width

```python
sns.FacetGrid(iris, hue='species', height=5).map(sns.distplot, 'petal_width').add_legend()
plt.show()
```

```python
# Yes this is good. We can see a considrable seperation of setosa from other two species.
# Also overlap is negotiable. 
# petal_width can be good choice to be included in model.
```

### petal_length

```python
sns.FacetGrid(iris, hue='species', height=5)\
   .map(sns.distplot, 'petal_length')\
   .add_legend()
plt.show()
```

```python
# This also a good choice.
# setosa is totally seperated from others.
# overlap is slightly less than petal_width. (Don't see the height of intersection see where the lines are starting)
```

### Conclusion

**Petal Length > Petal Width >>> Sepal Length >> Sepal Width**
