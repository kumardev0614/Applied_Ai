# K-S Test

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

### Generate a Normal random variable X

```python
x = stats.norm.rvs(size=1000)

sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

### Performing KS test on x to check it is normally distributed or not.

```python
stats.kstest(x, 'norm')
```

Here statistic represents our D(test-statistic) and pvalue is pvalue.

Here pvalue is very large, means we must accept our Null Hypotesis.
<br/>H0 = Distributions of both random variables are same. 

Means x is follows Normal distribution.

## KS test for Uniform Distribution.

```python
# y ~ continuous Uniform Dist(0,1)

y = np.random.uniform(0,1,10000)
# conti. uni. dist. between 0 and 1 of size 10000.

sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

### KS test on Y

```python
stats.kstest(y, 'norm')
```

pvalue is 0 means we have to reject our null hypo. Means Y does not follows Normal distribution.

### Check for Uniform Distribution.

```python
stats.kstest(y, 'uniform')
```

Here pvalue is 0.52 means 52%
<br/>We should accept H0. Means y follows Uniform Distribution.
