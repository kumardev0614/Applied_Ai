```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
m1 = 50
s1 = 15
```

```python
x = [m1-3*s1, m1-2*s1, m1-s1, m1, m1+s1, m1+2*s1, m1+3*s1]
```

```python
y = [55,23,14,12,52,14,31]
```

```python
print(x)
```

```python
sdx = []
for val in x:
    sdx.append((val - m1)/s1)
```

```python
print(sdx)
```

```python
plt.plot(x, y, 'r')
plt.show()
```

```python
plt.plot(sdx, y, 'g' )
plt.show()
```
