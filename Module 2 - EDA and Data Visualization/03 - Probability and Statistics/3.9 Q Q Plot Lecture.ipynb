# Q Q Plot

```python
import numpy as np
import pylab
import scipy.stats as stats
import seaborn as sns
```

### We can directly plot a Q Q Plot using scipy.stats

```python
# First we will check for Nrmally distributed variable

X_norm = np.random.normal(loc=20, scale=5, size=100)

# Here loc=mean=20 and scale=std=5
```

```python
sns.distplot(X_norm)
```

```python
# Q Q Plot

stats.probplot(X_norm, dist='norm', plot=pylab)
pylab.show()

# Here first argument is our random variable whose distribution is need to be compared.
# dist='norm' means comapre it with Standard Normal distribution.
```

```python
# As we can see we are getting a kind of straight line but it is not as good as we need.
# It does not mean that our X_norm is not Normal. 
# We can improve our results by increasing the size of X_norm.

X_norm_big = np.random.normal(loc=20, scale=5, size=1000)

stats.probplot(X_norm_big, dist='norm', plot=pylab)
pylab.show()

# Also Try size=10000
```

## Limitations of Q Q Plot

```python
# As we have observed Q Q plot performs poor in case of small size.
# So this is the limitation of Q Q plot.
```

# Q Q Plot for Non - Normal distribution.

```python
# Now we will create a Uniform random variable.

X_uni = np.random.uniform(low=-1, high=1, size=100)

sns.distplot(X_uni)
```

```python
# Q Q Plot

stats.probplot(X_uni, dist='norm', plot=pylab)
pylab.show()
```

```python
# It is confusing. Its hard to interpret the results. So we will increase the size.

X_uni_big = np.random.uniform(low=-1, high=1, size=1000)

stats.probplot(X_uni_big, dist='norm', plot=pylab)
pylab.show()
```

```python
# Now we can clearly see that the graph is not a straight line.
# It means X_uni_big does not follows Normal Distribution. So as X_uni
```
