## Creating Theoratical/Reference Dataset with Standard Normal Distribution.
Means mu = 0, and sigma = 1 (approx)

```python
mu = 0
sigma = 1
t = np.random.normal(mu, sigma, 1000)
```

```python
sns.distplot(t)
```

```python
print(np.mean(t))
print(np.std(t))
```

### This t will work as theoratical dataset. We will compare our datasets with this standard normal dataset.

```python
# We need to calculate it's percentile values also.
t_sort = sorted(t)
t_dash = np.percentile(t_sort, range(1,101))
```

```python
print(t_dash)
```
