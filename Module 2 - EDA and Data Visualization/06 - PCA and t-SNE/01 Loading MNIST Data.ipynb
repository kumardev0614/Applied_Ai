```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d0 = pd.read_csv("MNIST_train.csv")
```

```python
d0.head(5)
```

```python
d0.shape
```

```python
# We can see we have 42000 datapoints(images) and 785 columns.
# 784 columns contains image data of X matrix and 1 column is containing Y value which is label.
# But we need only X matric for now.
```

```python
# save label column in l, This l will work as Y(1xn) matrix.
l = d0['label']
```

```python
# Now make a dataframe with X values only.
X = d0.drop('label', axis=1)
```

```python
X.head(5)
```

```python
X.shape
```

# Accesing Image data using Matrix X

```python
# We will try to show the image by visualising any xi value in X.
# Because we have flatten out 28x28 matrix as xi in X.
# So to access a full image we have to convert flatten 784 dim xi, again into 28x28 matrix.

plt.figure(figsize = (7,7))
i = 35000

image_data = X.iloc[i].to_numpy().reshape(28,28)
# Take the ith xi from X. Convert it from Series to array. And reshape it's dimensions to 28x28.

plt.imshow(image_data, interpolation='none', cmap="gray")

# corresponding Y value
print("Y =", l[i])
```
