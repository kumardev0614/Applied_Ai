```python
# For column normalization we need min() and max() of each column.

def col_norm(column):
    c_min = min(column)
    c_max = max(column)
    col_n = []
    
    for value in column:
        col_n.append(((value - c_min) / (c_max - c_min)))   # formula for normalization
    
    return col_n
```

```python
col1 = data['sl']
col1
```

```python
f_dash  = col_norm(col1)
f_dash
```

```python
norm_matrix = pd.DataFrame(f_dash, columns =['sl_dash'])
norm_matrix
```

```python
col2 = data['sw']
f_dash  = col_norm(col2)

norm_matrix['sw_dash'] = f_dash
norm_matrix
```

```python
col3 = data['pl']
f_dash  = col_norm(col3)

norm_matrix['pl_dash'] = f_dash
norm_matrix
```

```python
col4 = data['pw']
f_dash  = col_norm(col4)

norm_matrix['pw_dash'] = f_dash
norm_matrix
```

```python
norm_matrix['species'] = data['species'] 
norm_matrix
```

<pre>So this is the normalised Matrix after applying column normalization on each column of original matrix.
Now for each column values are from only 0 to 1.
</pre>
