# HAVING

```python
# Print years which have >8000 movies in our DB [Data Scientist for Analysis]

query("SELECT year,COUNT(year) year_count FROM movies GROUP BY year HAVING year_count>8000;")
```

### Order of execution:
1. GROUP BY to create groups
2. apply the AGGREGATE FUNCTION
3. Apply HAVING condition.

### HAVING is often used along with GROUP BY. But it's Not Mandatory.

```python
query("SELECT name FROM movies WHERE year=2008;")
```

```python
query("SELECT name,year FROM movies HAVING year=2008;") # We have to incude year column in result to apply HAVING caluse on it
```

```python
# NOTE: WHERE works on each row of table and works before the GROUP BY clause. So using WHERE with group won't be possible.
# The rule of thumb is that use WHERE on columns before GROUP BY, use HAVING after GROUP BY mostly on aggrigated columns.
```

### WHERE and HAVING in same query

```python
# Q - Show the list of years when atleast 20+ movies with 9+ rating were released.

query("SELECT year, COUNT(year) movies_in_year FROM movies WHERE rankscore>9 GROUP BY year HAVING movies_in_year>20;")

# 1) First WHERE will be execute and we will get a table with 9+ rating movie in each row.
# 2) all these 9+ rating movies will be GROUPED BY year.
# 3) Then aggrigate function COUNT will calulate number of movies present in each year.
# 4) Then HAVING will select only those year which have movies_in_year > 20
```
