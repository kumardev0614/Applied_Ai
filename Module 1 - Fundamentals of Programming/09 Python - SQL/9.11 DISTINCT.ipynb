# DISTINCT

```python
# It is used to fetch unique values from a single column or multiple columns.
# When we are applying it to multiple columns then result of combined values of all columns should be unique.
```

### For Single Column

```python
pd.read_sql_query("SELECT * FROM movies_genres LIMIT 15", con)
```

```python
# We can see that for a single movie id there are multiple genres. But what if we want to see how many genres are 
# available in our dataset. # What are the all possible genres in our dataset which can be associate to a movie id.

# So here we will use DISTINCT.
```

```python
def query(q):
    return pd.read_sql_query(q, con)
```

```python
query("SELECT DISTINCT genre FROM movies_genres")
```

```python
# So there are total 20 distinct genre possible in our dataset.
```
