# WHERE

```python
# Where is mostly used to fetch data on certain condition.
```

```python
query("SELECT name,year,rankscore FROM movies WHERE rankscore>9;")
```

```python
# So here are 1069 rows in movies table where rankscore is greator than 9.

# Now find top 20 movies with rating > 9 also sort them in higher rating first.

query("SELECT name,year,rankscore FROM movies WHERE rankscore>9 ORDER BY rankscore DESC LIMIT 20;")
```

### NULL and WHERE clause
'=' does not work with NULL, It will give you an empty result-set.

```python
query("SELECT * FROM movies LIMIT 10")
```

```python
# We can see there are many NaN values in movies table so we will try to fetch those rows where rankscore is NULL.

query("SELECT * FROM movies WHERE rankscore=NULL")
```

```python
# We can see we got an empty record set. 
# let's run the same query using cursor.

c = con.cursor()

c.execute("SELECT * FROM movies WHERE rankscore=NULL")
c.fetchall()
```

### IS NULL, IS NOT NULL with WHERE

```python
# To compare null values we should use IS NULL and IS NOT NULL

query("SELECT * FROM movies WHERE rankscore IS NULL LIMIT 10")
```

```python
query("SELECT * FROM movies WHERE rankscore IS NOT NULL LIMIT 10")
```
