## Logical Operators
#### AND, OR, NOT, ALL, ANY, BETWEEN, EXISTS, IN, LIKE, SOME

### AND

```python
query("SELECT name,year,rankscore FROM movies WHERE rankscore>9 AND year>2000;")
```

### NOT

```python
query("SELECT name,year,rankscore FROM movies WHERE NOT year<=2000 LIMIT 20;")
```

### OR

```python
query("SELECT name,year,rankscore FROM movies WHERE rankscore>9 OR year>2007;")
```

### BETWEEN

```python
query("SELECT name,year,rankscore FROM movies WHERE year BETWEEN 2006 AND 2007;")

# Here values are inclusive in nature.
# Means year>=2006 AND year<=2007.
# Here 2006 is called lowvalue and 2007 is called highvalue
```

```python
query("SELECT name,year,rankscore FROM movies WHERE year BETWEEN 2007 AND 2006;")

# We cannot interchange values.
# Rule is lowvalue <= highvalue, else we will get an "empty" result-set. because year>=2007 AND year<=2006.
```

### LIKE
### % means 0 or more characters.
### _ means exactly 1 character. It has to be filled with a char.

```python
# pirates of %%%%%%%%%%%%%%%%%%%%

query("SELECT name,year FROM movies WHERE name LIKE 'pirates of %';")
```

```python
# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%dev

query("SELECT name,year FROM movies WHERE name LIKE '%dev';")
```

```python
# %%%%%%%%%%%%%%%of the%%%%%%%%%%%%%%%%%

query("SELECT name,year FROM movies WHERE name LIKE '%of the%';")
```

#### Examples for _ wildcard

```python
# suppose we have LIKE 'Agn_s'
# Accepted strings ==>  Agnes, Agnus, Agnos etc
# Non Acceptable strings ==> Agns, Agneus.  _ wilcard can't be empty also. While % wildcard can be.
```

```python
query("SELECT first_name, last_name FROM actors WHERE first_name LIKE 'Agn_s';")
```

```python
query("SELECT first_name, last_name FROM actors WHERE first_name LIKE 'jo_';")
```

```python
query("SELECT DISTINCT first_name FROM actors WHERE first_name LIKE 'jo_';")
```

### NOTE: If we want to find strings with % and _ as symbols then we have to use \\% and "\\_"
#### By using backslash they will be treated as symbols not as wildcards
