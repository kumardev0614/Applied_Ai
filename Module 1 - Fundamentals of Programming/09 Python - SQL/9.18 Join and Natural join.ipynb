```python
query("SELECT * FROM movies LIMIT 10")                # Overview of movies table
```

```python
query("SELECT * FROM movies_genres LIMIT 10")           # Overview of movies_genres table
```

### Show Name and Genre of movies.

```python
# Here names are in movies table and genres are in movies_genres table.
# But we have one thing common in both tables "id" in movies and "movie_id" in movies_genres table.
# So we can join these tables and get the corresponding genre to name.
```

### join all data in both tables by common columns.

```python
query("SELECT * FROM movies JOIN movies_genres ON movies.id=movies_genres.movie_id LIMIT 20;")
```

### Select only name and genre also use table name alias.

```python
query("SELECT m.name,g.genre FROM movies m JOIN movies_genres g ON m.id=g.movie_id LIMIT 20;")

# This is the same query as before but
# only m.name and g.genre columns are SELECTED
# m is used as alias for movies table, g is used as alias for movies_genres table.

# Order of execution
#  FROM and JOIN
#  SELECT 
#  LIMIT
```

# Natural Join

natural join: a join where we have the same column-names across two tables.

T1: C1, C2

T2: C1, C3, C4

SELECT * FROM T1 JOIN T2;

SELECT * FROM T1 JOIN T2 USING (C1);

returns C1,C2,C3,C4

no need to use the keyword "ON"
