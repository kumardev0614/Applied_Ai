### Import, connect, create cursor.

```python
import pandas as pd
import mysql.connector as sql

conn = sql.connect(host='localhost', user='root', password='mysql@dev', database='imdb')

c = conn.cursor()
```

```python
conn
```

## LIMIT
LIMIT is used to limit the number of rows fetched in result-set.

```python
c.execute("SELECT id,name,year FROM movies LIMIT 10;")
c.fetchall()
```

```python
pd.read_sql_query("SELECT id,name FROM movies LIMIT 8;", conn)
```

```python
pd.read_sql_query("SELECT id,name,year FROM movies LIMIT 8;", conn)
```

## OFFSET
It is used to skip the first n number of rows.

```python
# For example we can see the first 10 rows from previous query. We will skip first 5 rows using OFFSET.
```

```python
pd.read_sql_query("SELECT id,name,year FROM movies LIMIT 10 OFFSET 5;", conn)
```

## ORDER BY

```python
# ORDER BY year Ascending.

pd.read_sql_query('SELECT id,name,year FROM movies ORDER BY year LIMIT 10;',conn)
```

```python
# ORDER BY year Descending.

pd.read_sql_query('SELECT id,name,year FROM movies ORDER BY year DESC LIMIT 10;',conn)
```

```python
# Order by id descending

pd.read_sql_query('SELECT id,name,year FROM movies ORDER BY id DESC LIMIT 10;',conn)
```

```python
# order by rankscore descending

pd.read_sql_query('SELECT name,rankscore FROM movies ORDER BY rankscore DESC LIMIT 10;',conn)
```
